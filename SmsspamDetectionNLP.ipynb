```python
df2=pd.read_csv('/content/drive/MyDrive/Datasets/SMS_test.csv',encoding='ISO-8859-1')
df2
```

```python
#combining two df

df=pd.concat([df1,df2],axis=0,ignore_index=True)
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
df
```

```python
df['Label']=df['Label'].map({'Non-Spam':0,'Spam':1})
```

```python
text=df['Message_body']
text
```

```python
#Remove special characters
text= text.str.replace('[^a-zA-z0-9]+'," ")
text
```

```python
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize

import nltk
nltk.download('punkt')

stemmer=SnowballStemmer('english')

text=text.apply(lambda x:[stemmer.stem(token.lower()) for token in word_tokenize(x)]).apply(lambda token:' '.join(token))
text
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw=stopwords.words('english')
```

```python

text=text.apply(lambda x:[stemmer.stem(token.lower()) for token in word_tokenize(x) if token not in sw]).apply(lambda token:' '.join(token))

text
```

```python
text=text.apply(lambda x:[stemmer.stem(token.lower()) for token in word_tokenize(x) if len(token)>=3]).apply(lambda token:' '.join(token))

text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(text)
print(X)
```

```python
X.toarray()
```

```python
y=df['Label']
y
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3,random_state=1)

X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report


knn=KNeighborsClassifier()
sv=SVC()
randomf=RandomForestClassifier()

models=[knn,sv,randomf]

for model in models:
  print("*************",model,"************")
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print(classification_report(y_test,y_pred))
  nw=vectorizer.transform([' guy bitch act like interest buy someth next '])
  nw=nw.toarray()
  y=model.predict(nw)
  if y==0:
    print('Non spam')
  else:
      print('spam')
```
