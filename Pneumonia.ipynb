```python
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
SIZE=[224,224] #this is the size used for vgg16
```

```python
train_path='/content/drive/MyDrive/Pneumonia/train'
test_path= '/content/drive/MyDrive/Pneumonia/test'
```

```python
#include top is added to give fully connected layer if we need give it as True or False
vgg=VGG16(input_shape=SIZE+[3],weights='imagenet',include_top=False)

#3 is for rgb, weights are already defined, it is available at imagenet
#if we are using only convolution and pooling layer then false, if we need fully connected layer also then true.
```

```python
#if we are using a pretrained model use this code to prevent  further tain of model with our data

for layer in vgg.layers:
  layer.trainable=False
```

```python
#if we are using a pretrained model use this code to prevent  further tain of model with our data

for layer in vgg.layers:
  layer.trainable=False
```

```python
x=Flatten()(vgg.output)
prediction=Dense(2,activation='softmax')(x)
```

```python
model=Model(inputs=vgg.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Image augmentation

train_datagen=ImageDataGenerator(rescale=1/255,
                                 zoom_range=0.2,
                                 horizontal_flip=True)

test_datagen=ImageDataGenerator(rescale=1/255)


training_set=train_datagen.flow_from_directory(train_path,
                                               target_size=(224,224),
                                               batch_size=32,
                                               class_mode='categorical')
```

```python
testing_set=test_datagen.flow_from_directory(test_path,
                                               target_size=(224,224),
                                               batch_size=32,
                                               class_mode='categorical')
```

```python
validation_path='/content/drive/MyDrive/Pneumonia/val'
```

```python
Validation_set=test_datagen.flow_from_directory(validation_path,
                                               target_size=(224,224),
                                               batch_size=32,
                                               class_mode='categorical')
```

```python
model.fit(training_set,validation_data=Validation_set,epochs=10)
```

```python
loss,accuracy=model.evaluate(testing_set)
print(loss)
print(accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score

# Step 1: Predict on the validation set
y_pred_val = model.predict(Validation_set)
y_pred_val_classes = np.argmax(y_pred_val, axis=1)  # Convert predicted probabilities to class labels

# Step 2: Extract true labels from the validation set
y_true_val = Validation_set.classes

# Step 3: Display confusion matrix for the validation set
ConfusionMatrixDisplay.from_predictions(y_true_val, y_pred_val_classes)
plt.title('Confusion Matrix - Validation Set')
plt.show()

# Step 4: Print classification report (Precision, Recall, F1-score, Accuracy)
print("Classification Report - Validation Set:\n", classification_report(y_true_val, y_pred_val_classes))

# Step 5: Compute and print overall accuracy
accuracy_val = accuracy_score(y_true_val, y_pred_val_classes)
print(f"Validation Set Accuracy: {accuracy_val:.4f}")


# Repeat the same process for the testing set

# Step 6: Predict on the testing set
y_pred_test = model.predict(testing_set)
y_pred_test_classes = np.argmax(y_pred_test, axis=1)  # Convert predicted probabilities to class labels

# Step 7: Extract true labels from the testing set
y_true_test = testing_set.classes

# Step 8: Display confusion matrix for the testing set
ConfusionMatrixDisplay.from_predictions(y_true_test, y_pred_test_classes)
plt.title('Confusion Matrix - Testing Set')
plt.show()

# Step 9: Print classification report (Precision, Recall, F1-score, Accuracy)
print("Classification Report - Testing Set:\n", classification_report(y_true_test, y_pred_test_classes))

# Step 10: Compute and print overall accuracy
accuracy_test = accuracy_score(y_true_test, y_pred_test_classes)
print(f"Testing Set Accuracy: {accuracy_test:.4f}")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
import numpy as np

# Step 1: Predict class labels for the training set
y_pred_train = model.predict(training_set)
y_pred_train_classes = np.argmax(y_pred_train, axis=1)  # Convert probabilities to class labels

# Step 2: Obtain true labels from the training set
y_true_train = training_set.classes

# Step 3: Plot the confusion matrix for the training set
ConfusionMatrixDisplay.from_predictions(y_true_train, y_pred_train_classes)
plt.title('Confusion Matrix - Training Set')
plt.show()

# Step 4: Generate the classification report (Precision, Recall, F1-score, etc.)
print("Classification Report for Training Set:\n", classification_report(y_true_train, y_pred_train_classes))

# Step 5: Print the overall accuracy
accuracy_train = accuracy_score(y_true_train, y_pred_train_classes)
print(f"Accuracy for Training Set: {accuracy_train:.4f}")
```
